```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    "Category": ["A", "B", "C", "D", "E"],
    "Values": [10, 25, 15, 30, 20]
}

df = pd.DataFrame(data)

plt.figure()
plt.bar(df["Category"], df["Values"])
plt.title("Bar Chart")
plt.show()

plt.figure()
plt.pie(df["Values"], labels=df["Category"], autopct="%1.1f%%")
plt.title("Pie Chart")
plt.show()

plt.figure()
plt.hist(df["Values"], bins=5)
plt.title("Histogram")
plt.show()
```
