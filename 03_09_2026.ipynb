```python
from sklearn.linear_model import LinearRegression
import pandas as pd

data = {
    "Size": [500, 800, 1000, 1200, 1500],
    "Price": [100, 150, 200, 230, 300]
}

df = pd.DataFrame(data)

X = df[["Size"]]
y = df["Price"]

model = LinearRegression()
model.fit(X, y)

predicted = model.predict([[1100]])
print(predicted)
```
