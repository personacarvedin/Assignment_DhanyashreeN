```python
# Creating 5 sample documents

doc1 = "Machine learning helps computers learn from data and improve automatically."

doc2 = "Artificial intelligence and machine learning are important technologies in modern software."

doc3 = "Students use machine learning for projects like prediction and classification tasks."

doc4 = "Data science includes statistics machine learning and data visualization techniques."

doc5 = "Python programming is widely used for machine learning and artificial intelligence applications."

documents = [doc1, doc2, doc3, doc4, doc5]

print("Documents created successfully!")
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd

# Apply TF-IDF

vectorizer = TfidfVectorizer(stop_words='english')

tfidf_matrix = vectorizer.fit_transform(documents)

feature_names = vectorizer.get_feature_names_out()

df = pd.DataFrame(tfidf_matrix.toarray(), columns=feature_names)

print("TF-IDF Table:\n")
print(df)

# Finding top 3 important words from each document

print("\nTop Keywords from Each Document:\n")

for i in range(len(documents)):

    top_words = df.iloc[i].sort_values(ascending=False).head(3)

    print(f"Document {i+1} Top Words:")
    print(list(top_words.index))
    print()
```
