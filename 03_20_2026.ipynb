```python
import pandas as pd
import re

# Step 1: Create 20 messy sentences dataset

sentences = [
    "Hey broo how r u?? 😂",
    "I cant cm to clg today",
    "This movie is sooo gud!!!",
    "LOL that was funny 🤣",
    "Plz send me the notes asap",
    "I didnt undrstand the lecture",
    "OMG exam is tmrw 😭",
    "Lets meet @ 5pm",
    "Thx for ur help",
    "Wat r u doing now??",
    "This is amazng 🔥",
    "Cant wait fr holidays!!!",
    "Gud mrng everyone ☀",
    "Study hardd guys",
    "IDK what to do",
    "Pls share the assignmnt",
    "That party was lit 😎",
    "Sry I missed the class",
    "BTW where r u?",
    "Im feeling tiredd today"
]

# Step 2: Define slang words list

slang_words = ["lol","omg","idk","btw","pls","plz","thx","ur","r","tmrw","cm","fr","sry","gud","broo"]

# Step 3: Emoji detection pattern

emoji_pattern = re.compile("["
                           u"\U0001F600-\U0001F64F"
                           u"\U0001F300-\U0001F5FF"
                           u"\U0001F680-\U0001F6FF"
                           u"\U0001F1E0-\U0001F1FF"
                           "]+", flags=re.UNICODE)

# Step 4: Simple typo detection (double letters example)

def detect_typos(text):
    return bool(re.search(r'(.)\1{2,}', text))

# Step 5: Process sentences

results = []

for sentence in sentences:

    slang_found = [word for word in sentence.lower().split() if word in slang_words]

    emoji_found = emoji_pattern.findall(sentence)

    typo_found = detect_typos(sentence)

    preprocessing_needed = []

    if slang_found:
        preprocessing_needed.append("Replace slang words")

    if emoji_found:
        preprocessing_needed.append("Remove emojis")

    if typo_found:
        preprocessing_needed.append("Correct spelling errors")

    preprocessing_needed.append("Lowercase conversion")
    preprocessing_needed.append("Remove punctuation")

    results.append([
        sentence,
        slang_found,
        emoji_found,
        typo_found,
        ", ".join(preprocessing_needed)
    ])

# Step 6: Create dataframe

df = pd.DataFrame(results, columns=[
    "Sentence",
    "Slang Found",
    "Emoji Found",
    "Typo Found",
    "Preprocessing Needed"
])

# Step 7: Show output

print("Messy Sentences Analysis:\n")
print(df)
```
