```python
import pandas as pd

# Create messy dataset
data = {
    "Name": ["Amit", "Riya", "Karan", "Sneha", "Rahul", "Amit"],
    "Age": [21, None, 20, 23, 21, 21],
    "Marks": [85, 90, None, 92, 88, 85],
    "City": ["bangalore", "Mumbai", "delhi", "CHENNAI", None, "bangalore"]
}

df = pd.DataFrame(data)

# Save dataset
df.to_csv("students_dirty.csv", index=False)

print("Dataset created successfully!")
print(df)
```

```python
import pandas as pd

# Load dataset
df = pd.read_csv("students_dirty.csv")

# Handle missing values
df["Age"] = df["Age"].fillna(df["Age"].mean())
df["Marks"] = df["Marks"].fillna(df["Marks"].mean())
df["City"] = df["City"].fillna("Unknown")

# Remove duplicates
df = df.drop_duplicates()

# Standardize text formatting
df["City"] = df["City"].str.title()

# Display cleaned dataset
print(df)

# Save cleaned dataset
df.to_csv("students_cleaned.csv", index=False)
```
