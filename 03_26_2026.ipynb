```python
# Step 1: Import library
from textblob import TextBlob

# Step 2: Create 5 movie reviews

reviews = [
    "The movie was amazing and I loved the acting",
    "The film was boring and too long",
    "Excellent story and wonderful direction",
    "The movie was not interesting and very slow",
    "Fantastic visuals and great performance by actors"
]

# Step 3: Analyze sentiment

print("Movie Review Sentiment Analyzer\n")

for review in reviews:

    sentiment = TextBlob(review).sentiment.polarity

    if sentiment > 0:
        result = "Positive Review 😊"

    elif sentiment < 0:
        result = "Negative Review 😞"

    else:
        result = "Neutral Review 😐"

    print("Review:", review)
    print("Sentiment Score:", sentiment)
    print("Result:", result)
    print("-----------------------------")
```
