```python
# Step 1: Import libraries
import nltk
from nltk.corpus import wordnet as wn

# Step 2: Download WordNet dataset
nltk.download('wordnet')

# Step 3: Create 5 word pairs

word_pairs = [
    ("car", "automobile"),
    ("student", "learner"),
    ("book", "pen"),
    ("teacher", "professor"),
    ("dog", "cat")
]

# Step 4: Function to calculate similarity

def get_similarity(word1, word2):

    synsets1 = wn.synsets(word1)
    synsets2 = wn.synsets(word2)

    if synsets1 and synsets2:
        similarity = synsets1[0].wup_similarity(synsets2[0])
        return similarity
    else:
        return None


# Step 5: Print similarity results

print("Semantic Similarity Between Word Pairs:\n")

for word1, word2 in word_pairs:

    similarity = get_similarity(word1, word2)

    print(f"{word1}  <-->  {word2}")
    print("Similarity Score:", similarity)
    print("-----------------------------")
```
