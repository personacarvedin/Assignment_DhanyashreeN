```python
# Step 1: Import libraries
import string
import nltk

# Step 2: Download stopwords
nltk.download('stopwords')

from nltk.corpus import stopwords

# Step 3: Sample messy text dataset

sentences = [
    "Hello!!! How are YOU doing today?",
    "This is a SIMPLE text-cleaning example.",
    "Python is VERY useful for NLP tasks.",
    "Please remove the stopwords from THIS sentence.",
    "Machine Learning is FUN and powerful!!!"
]

# Step 4: Load stopwords list
stop_words = set(stopwords.words('english'))

# Step 5: Create text cleaning function

def clean_text(text):

    # Convert to lowercase
    text = text.lower()

    # Remove punctuation
    text = text.translate(str.maketrans('', '', string.punctuation))

    # Remove stopwords
    words = text.split()
    cleaned_words = [word for word in words if word not in stop_words]

    # Join words back into sentence
    cleaned_text = " ".join(cleaned_words)

    return cleaned_text


# Step 6: Apply cleaning function

print("Original Text  --->  Cleaned Text\n")

for sentence in sentences:
    cleaned = clean_text(sentence)
    print(sentence, " ---> ", cleaned)
```
