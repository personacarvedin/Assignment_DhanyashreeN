```python
import pandas as pd

# Create dataset
data = {
    "Name": ["Amit", "Riya", "Karan", "Sneha", "Rahul"],
    "Age": [21, 22, 20, 23, 21],
    "Marks": [85, 90, 78, 92, 88],
    "City": ["Bangalore", "Mumbai", "Delhi", "Chennai", "Hyderabad"]
}

df = pd.DataFrame(data)

# Save as CSV file
df.to_csv("students.csv", index=False)

print("CSV file created successfully!")
```

```python
import pandas as pd

df = pd.read_csv("students.csv")

print("Top rows:")
print(df.head())

print("\nHighest values:")
print(df.max(numeric_only=True))

print("\nColumn with highest value:")
print(df.max(numeric_only=True).idxmax())

print("\nMissing values:")
print(df.isnull().sum())
```
